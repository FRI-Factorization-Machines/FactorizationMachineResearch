```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import category_encoders as ce
```

```python
company_data = pd.read_csv("CompanyData.csv")
company_data.drop(1760, axis=0, inplace=True)
company_data.drop(['Unnamed: 15', ' City', ' Phone', ' Fax', ' Zip', ' Zip2', ' Plugger', ' Driller', ' Financial Security Provider', ' Address 1', ' Address 2'], axis=1, inplace=True)
```

```python
# encodes a dataframe given a column name
def encoding(df, col_name):
    for col in df.columns:
        if col != col_name:
            df = df.drop(col, axis = 1)
    encoder=ce.OneHotEncoder(cols=df.columns[0],handle_unknown='return_nan',return_df=True,use_cat_names=True)
    data_encoded = encoder.fit_transform(df)
    return data_encoded
```

```python
# encodes a one column dataframe
def simple_encoding(df):
    encoder=ce.OneHotEncoder(cols=df.columns[0],handle_unknown='return_nan',return_df=True,use_cat_names=True)
    data_encoded = encoder.fit_transform(df)
    return data_encoded
```

```python
encoding(company_data, " State")
```
