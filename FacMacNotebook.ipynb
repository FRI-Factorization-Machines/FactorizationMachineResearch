```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from rankfm.rankfm import RankFM
```

```python
well_data = pd.read_csv("wellspublic.csv")
```

```python
transfer_data = pd.read_csv("WellTransferData.csv")
```

```python
well_data
```

```python
input_data = well_data[[ "Operator_number","API_WellNo",]]
```

```python
input_data
```

```python
from rankfm.rankfm import RankFM
model = RankFM(factors=20, loss='warp', max_samples=20, alpha=0.01, sigma=0.1, learning_rate=0.1, learning_schedule='invscaling')
model.fit(input_data, epochs=20, verbose=True)
```

```python
valid_scores = model.predict(input_data, cold_start='nan')
```

```python
valid_scores
```

```python
valid_recs = model.recommend(input_data["Operator_number"][:5], n_items=10, filter_previous=True, cold_start='drop')
```

```python
valid_recs
```

```python
from rankfm.evaluation import hit_rate, precision, recall

model_hrt = hit_rate(model, input_data, k=10)
model_pre = precision(model, input_data, k=10)
model_rec = recall(model, input_data, k=10)
```

```python
model_hrt, model_pre, model_rec
```
