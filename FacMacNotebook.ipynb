```python
from rankfm.rankfm import RankFM
model = RankFM(factors=20, loss='warp', max_samples=20, alpha=0.01, sigma=0.1, learning_rate=0.1, learning_schedule='invscaling')
model.fit(interactions=interactions[5:], item_features=item_features[5:], sample_weight = np.ones(len(interactions)-5)/(len(interactions)-5),epochs=20, verbose=True)
```
