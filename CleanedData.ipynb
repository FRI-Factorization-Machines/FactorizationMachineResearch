```python
#manually rename rows to eliminate possible spacing issues
wells_data.columns = ['API_WellNo', 'Well Name', 'Company_name', 'Operator_number', 'Date_Well_Completed', 'Date_well_plugged']
wells_data.dropna(subset=['API_WellNo', 'Well Name', 'Operator_number', 'Date_Well_Completed'], inplace=True)
wells_data
```

```python
#reformat dates based on century
from datetime import datetime
import datetime as dt
def fix_date(x):

    if x.year > 2021:

        year = x.year - 100

    else:

        year = x.year

    return dt.date(year,x.month,x.day)
```

```python
comp_arr = []
plug_arr = []
for index, row in wells_data.iterrows():
    if pd.isna(row["Date_well_plugged"]):
        datetime_wellplugobj = datetime.now()
    else:
        datetime_wellplugobj = datetime.strptime(row["Date_well_plugged"], '%Y-%m-%d %H:%M:%S')

    datetime_wellcompobj = datetime.strptime(row["Date_Well_Completed"], '%Y-%m-%d %H:%M:%S')
    comp_arr.append(datetime_wellcompobj)
    plug_arr.append(datetime_wellplugobj)
```

```python
wells_data['Comp_date'] = comp_arr
wells_data['Plug_date'] = plug_arr
```

```python
wells_data['Comp_date'] = wells_data['Comp_date'].apply(fix_date)
wells_data['Plug_date'] = wells_data['Plug_date'].apply(fix_date)
```

```python
diff_arr = []
for index, row in wells_data.iterrows():
    diff_arr.append((row['Plug_date'] - row['Comp_date']).days)
wells_data['Num Days'] = diff_arr
```
